```python
import pandas as pd

# Load just one of the files first — it's smaller and faster to iterate
df = pd.read_csv("data/raw/USHY_INDEX_20260301_part_1.csv")

# Check 1: shape and columns
print(df.shape)         # rows × columns
print(df.columns.tolist())
df.head()

# Check 2: what are the date types?
df["Date"].head()       # Are they strings? "1/1/2010" format?
df["Date"] = pd.to_datetime(df["Date"])
print(df["Date"].min(), df["Date"].max())  # Date range
print(df["Date"].nunique(), "unique dates")  # How many months?

# Check 3: how many bonds per month?
bonds_per_month = df.groupby("Date")["Cusip"].nunique()
print(bonds_per_month.describe())
bonds_per_month.plot(title="Bonds in index per month")

# Check 4: rating distribution
print(df["Eff_Rating_Group"].value_counts())

# Check 5: sector distribution
print(df["Class3"].value_counts())

# Check 6: do return columns look like percent or decimal?
print(df["Total_Return_MTD"].describe())
# If you see values like 6.41, -8.25 — those are PERCENT (e.g., +6.41%)
# If you see values like 0.0641, -0.0825 — those are decimals
# This matters: you need to decide once, at ingest, and stick with it.

# Check 7: OAS distribution by rating (should rise BB → B → CCC)
df.groupby("Eff_Rating_Group")["OAS"].median()
# Expect something like: BB ~300, B ~500, CCC ~1000+

# Check 8: missing data
print(df.isna().sum())
# Some columns will have NaNs — note which ones.

# Check 9: now load BOTH files and verify they stitch together cleanly
df1 = pd.read_csv("data/raw/USHY_INDEX_20260301_part_1.csv")
df2 = pd.read_csv("data/raw/USHY_INDEX_20260301_part_2.csv")
df_full = pd.concat([df1, df2], ignore_index=True)
df_full["Date"] = pd.to_datetime(df_full["Date"])
print(df_full["Date"].min(), df_full["Date"].max())
# Verify the date range goes from Jan 2010 all the way to March 2026.
# Verify there's no gap or overlap between the two files.
```
